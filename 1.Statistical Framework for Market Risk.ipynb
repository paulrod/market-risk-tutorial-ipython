### Modeling Stock Price Evolution as IID

We wish to build a model that describes the evolution of the stock price through time.

Consider a set of random variables that represent the stock price through time. If we begin with the assumption that the stock price is consistent with the Efficient Market Hypothesis (EMH), then the price at any time incorporates all possible information about the stock. This means that there is no additional information to be gained by looking at prices prior to the current one, or that the random variables representing the price are independent of one another. 

The EMH is typically aligned with the assumption that the random variables that represent our stock price are Independent and Identically Distributed (IID) through time. IID means not only that that each variable is independent of any prior one, but also that each variable has an identical probability distribution. IID is one of the simplest and strongest assumptions we can make about the random variables and if we find that it is true then our modeling task will be simplified. 

We'll use IID as the base assumption for our model and relax it later as necessary. To get a gneral sense whether stock prices can be modeled as IID we'll use two quick visual tests:

- Independence: Make a scatter plot of the prices vs the one period lagged price. There should be no relation between the two time series because each price is independent of the prior one. The Autocorrelation Function (ACF) extends this test to multiple lag periods. 

- Identically Distributed: Divide the time series into 2 halves and graph the histograms of each, they should be visually similar. 

### Stock Prices
We first test the historical stock prices to determine if they align with the IID assumption.

```python
# Histograms of the AAPL time series halves
plt.subplot(121)
tmp = px_adj_close['AAPL'][:len(px_adj_close['AAPL'])/2].hist(bins=80, figsize=(14,6))

plt.subplot(122)
px_adj_close['AAPL'][len(px_adj_close['AAPL'])/2:].hist(bins=80)
x_ = plt.ylim(tmp.get_ylim())
```

The stock price time series halves do not look similar.

We now test for independence of the stock prices. 

```python
# Testing for independence at lag 1 and the ACF
fig, ax = plt.subplots(1, 2, figsize=(14,6))
lag_plot(px_adj_close['AAPL'], lag=1, ax=ax[0])
autocorrelation_plot(px_adj_close['AAPL'], ax=ax[1])
```

The lag 1 plot and ACF show us that rather than being independent, prices are almost perfectly autocorrelated at lag 1 with a decaying autocorrelation as the lag increases. 

So stock prices are not IID. To examine this in more detail we look at the Partial Autocorrelaiton Function (PACF) to determine the extent of the autocorrelation. 

```python
# PACF from Statsmodels
x_ = sm_graphics.tsa.plot_pacf(px_adj_close['AAPL'], lags=20)
```

The PACF shows us that there is no remaining autorrelation after removing that present at lag 1. 

This indicates that the stock price at time $t$ is heavily dependent on the stock price at $t-1$, but not on any prices prior to that. So while stock prices are not IID, they do have the Markov property of memorylessness, i.e. where the next step is not dependent on anything other than the current step. This finding does not contradict the EMH, but rather illustrates an important characteristic of the time series evolution of the stock price itself. We can model this evolution using a lag 1 autoregressive AR(1) type model. 

$$P_t = P_{t-1} + \epsilon_t,\ \epsilon_t \sim IID\ N(0,1)$$

Here the $P_{t-1}$ term represents the observed lag 1 autocorrelation and the stochastic term $\epsilon_t$ generates IID samples from a standard normal distribution, representing the uncertainty in the next day's price evolution. 

With this model, the lag 1 autocorrelation in the price serves to establish a general price level for time $t$, to which we add an independent random change to get the day's price $P_t$.

A price that follows a path generated by a series of IID changes is said to follow a Random Walk (or a Wiener Process/Brownian Motion in continuous time). 

Let's simulate the random walk and see if it compares to the actual stock price evolution. 

```python
# Modeling the random walk
start_price = px_adj_close['AAPL'][0]
n = len(px_adj_close['AAPL'])
dates = px_adj_close['AAPL'].index

# Build the model P(t) = P(t-1) + e(t)
change_model = np.zeros(n)
change_model[0] = start_price # P0
change_model[1:] = np.random.randn(n-1) # the stochastic component
price_model = change_model.cumsum()

# plot AAPl prices and our model - try this multiple times to understand the model
plt.figure(figsize=(14,5))
for y, label in zip([px_adj_close['AAPL'].values, price_model], ['AAPL', 'model']):
    plt.plot(dates, y, label=label)
plt.grid(True)
plt.legend(loc='upper left')
```

If we simulate the random walk model mutliple times we see that it does not replicate the upward trend of the AAPL stock price.

To model this trend we add in a fixed parameter $\mu$ called the drift, which represents a constant per time step increase in the stock price. We'll also add in a scaling factor $\sigma$ so that we can control the scale of the stochastic term generating the daily price changes.

$$P_t = \mu + P_{t-1} + \sigma\epsilon_t$$

We call this the random walk with drift and it has a conditional expectation and variance which represent the price $P_t$ at time $t$ conditional on an initally observed price $P_0$. We can iteratively substitute the process equation into itself to arrive at these values:

$E[P_t|P_0] = P_0 + \mu{t}$

$Var[P_t|P_0] = \sigma^2t$

So at time $t$ we expect the price to be the sum of the initial price $P_0$ and $\mu{t}$ increments of drift. We expect the variance of the process at time $t$ to be $\sigma^2t$, and the std dev to be $\sigma\sqrt{t}$.

```python
# We can back out the drift and scale parameters of our random walk model for AAPL
mu =(px_adj_close['AAPL'][-1] - px_adj_close['AAPL'][0])/len(px_adj_close['AAPL']) # (P(t) - P(0)/t
sigma = px_adj_close['AAPL'].std()/np.sqrt(len(px_adj_close['AAPL'])) # sigma(t)/sqrt(t)
mu, sigma
```

```python
# General function for an AR(1) process with drift mu, std dev sigma and a lag term coefficient c
def ar1(n, init=0, mu=0, sigma=1, c=1):
    ts = np.zeros(n)
    ts[0] = init
    for i in range(n-1):
        ts[i+1] = mu + c*ts[i] + sigma*np.random.randn()
    return ts

# simulate a random walk with drift for AAPL
price_model = ar1(n=len(px_adj_close['AAPL']), init=px_adj_close['AAPL'][0], mu=mu, sigma=sigma) 

plt.figure(figsize=(14,5))
for y, label in zip([px_adj_close['AAPL'].values, price_model], ['AAPL Price', 'AR(1) Model Price']):
    plt.plot(px_adj_close['AAPL'].index, y, label=label)
plt.grid(True)
plt.legend(loc='upper left')
```

Our random walk model with drift now more closely replicates the upward trend of the actual AAPL prices.

### Stock Price Differences

Earlier we noted that we wished to model our stock as a series of IID random variables. We now know that those random variables should not be the stock prices themselves. A more appropriate candidate would be the price changes that cumulatively describe the stock price at some time $t$. Per the random walk model we already know that these price changes are IID samples from a normal distribution ~$N(\mu,\sigma^2)$, so all that's left is to test whether this holds up vs. actual prices. 

We can modify our random walk model to get these stock price differences $D_t$.

$$D_t = P_t - P_{t-1} = \mu + \sigma\epsilon_t$$

Let's perform the IID checks on the stock price differences. 

```python
# Function to check whether the two halves of a time series have identical distributions
def plot_histogram_halves(timeseries, nbins):
    figure = plt.figure(figsize=(8,6))
    x1, x2 = np.array_split(timeseries,2) # array_split is OK with an odd length series

    # plot the normed histograms
    n,bins,patches=plt.hist(x1, bins=nbins, normed=True, alpha=.5, color='r', label='first half')
    # print np.sum(n*np.diff(bins)) #Prove your hist integrates to 1
    plt.hist(x2, bins=nbins, normed=True, alpha=.5, color='b', label='second half')

    plt.grid(True)
    plt.legend()

    
# Check identicalness for IBM
px_adj_close['IBM diff'] = px_adj_close['IBM'] - px_adj_close['IBM'].shift(1)
plot_histogram_halves(px_adj_close['IBM diff'].dropna(),50)
plt.title('IBM Timeseries Halves')
x_ = plt.xlim(-6,6)
```

```python
# Testing for independence 
fig, ax = plt.subplots(1, 2, figsize=(14,6))
lag_plot(px_adj_close['IBM diff'].dropna(), ax=ax[0])
autocorrelation_plot(px_adj_close['IBM diff'].dropna(), ax=ax[1])
```

Price differences are independent, as shown by the lag 1 plot and ACF. The histograms show us that the IBM time series halves are not exactly the same, but they do have the same general shape and may still be OK for us even though not strictly from identical distributions.

Let's test the identicalness of the AAPL price differences to confirm if this is a reasonable assumption.

```python
px_adj_close['AAPL diff'] = px_adj_close['AAPL'] - px_adj_close['AAPL'].shift(1)

plot_histogram_halves(px_adj_close['AAPL diff'].dropna(), 40)
plt.xlim(-3,3)
```

We can no longer say that the histogram halves for AAPL are reasonably similar. The second half has a much larger variance than the first. So stock price differences are not IID. 

Plotting the time series for the differences directly, we can see the changing variance over time. 

```python
# We can see the increasing variance of AAPL in the time series plot
px_adj_close['AAPL diff'].plot(figsize=(14,5))
```

This non-constant variance is partly due to the fact that the magnitude of the difference at time $t$ is relative to the stock price level at time $t$. 

For example, consider two similar stocks priced at \$100 and \$30 per share, respectively. If the margins for the market they are in improve by 10% we might expect their prices to increase by a similar amount, which would be \$10 and \$3 per share, respectively. The stock with the $10 increase would reflect a higher variance in price differences, which is not what we want to measure. This same example could be applied to a single stock with different price levels during different pricing regimes over time. 

## Percentage Return

As suggested above, we want to measure the relative, rather than the absolute changes in the stock price. This is described by percentage returns, which are defined as:

Simple, Linear or Net Return
$$R_t = \frac{P_t - P_{t-1}}{P_{t-1}} = \frac{P_t}{P_{t-1}}-1$$

Total or Gross Return
$$(1 + R_t) = \frac{P_t}{P_{t-1}}$$

Simple return is what is quoted when discussing stock returns.

```python
# Calculating single period return
Pprev = px_adj_close['AAPL'].loc['20150106']
Pcurr = px_adj_close['AAPL'].loc['20150107']

print 'Simple Return: {:.2%}'.format((Pcurr-Pprev)/Pprev)
print 'Simple Return: {:.2%}'.format(Pcurr/Pprev - 1)
print 'Total Return: {:.2%}'.format(Pcurr/Pprev)
```

### Stock Returns
The process for stock price changes written in terms of total return is then: 

$$(1+R_t) = \frac{P_t}{P_{t-1}} = \mu + \sigma\epsilon_t$$

Which we note has become multiplicative, rather than cumulative. 

```python
# Calculating the total returns
returns = pd.DataFrame()

returns['AAPL Total Rtn'] = (px_adj_close['AAPL']/px_adj_close['AAPL'].shift(1)).dropna() # the first rtn is N/A, drop it
returns.head()
```

We perform the IID tests on the total returns

```python
# Testing AAPL for independence 
fig, ax = plt.subplots(1,2, figsize=(14,6))

lag_plot(returns['AAPL Total Rtn'], ax=ax[0])
autocorrelation_plot(returns['AAPL Total Rtn'], ax=ax[1])
```

```python
# AAPL total returns
plot_histogram_halves(returns['AAPL Total Rtn'], 40)
```

```python
# IBM total returns
returns['IBM Total Rtn'] = (px_adj_close['IBM']/px_adj_close['IBM'].shift(1)).dropna()
plot_histogram_halves(returns['IBM Total Rtn'], 40)
```

Total returns are independent and the time series halves now look much more more identical than they did for price differences. 

We can also observe the time series plots of the total returns.

### Non-Constant Variance 

Using returns to describe the relative changes in the stock price we still observe non-constant variance in the time series for both stocks. In particular, we note a sustained period of high variance during the 2008-2009 financial crisis. Non-constant variance is a known feature of financial time series. Variance tends to spike in times of market stress and holds sustained high or low levels with a gradual change back to the long run average. 

We have multiple options for modeling the evolution of the variance, which we can use in the random walk model for returns. We won't go into them here, but we note that this would change the $\sigma$ term in the random walk to $\sigma_t$. 

$$(1+R_t) = \mu + \sigma_t\epsilon_t$$

### Market Invariant for a Stock

We here introduce the term Market Invariant for a stock, which is a convenient way to describe the series of random variables that we have been searching for. Our market invariant for a stock thus far is the total return, and we'll be satisfied that this market invariant approximates the IID behavior we're after. Invariant means unchanging and this is the property we're after for our stock. If we can model our stock with a market invariant based on observed historical prices, we can then forecast the evolution of the market invariant to unobserved prices. 

Simple returns are a linear transform of the total returns, and as such they are an equivalently suitable market invariant.

## Properties of Returns

Now that we have returns as a market invariant, we need to understand two important properties of working with them:

- Aggregating returns temporally (across time)
- Aggregating returns cross-sectionally (across a portfolio of stocks)

### Aggregating returns temporally

Given $n$ simple returns, their discretely compoundeded return over time $R_t(n)$ is the product of the $n$ total returns.

$$(1 + R_t(n)) = \frac{P_t}{P_{t-n}} = \Pi(1 + R_{t-n}) = (1 + R_t)*(1 + R_{t-1})*\ldots*(1 + R_{t-n+1})$$

Note how the aggregate total return over the period can be determined either by the cumulative product of the single period returns, or by using only the most recent price $P_t$ and the initial price $P_{t-n}$. The discretely compoundeded simple return is found by subtracting 1:

$$R_t(n) = \frac{P_t}{P_{t-n}}-1 = \Pi(1 + R_{t-n}) -1 = [(1 + R_t)*(1 + R_{t-1})*\ldots*(1 + R_{t-n+1})] - 1$$

##### Quoting return in the market (Aside)
We often wish to quote the average of the $n$ returns. This is done with the geometric mean, which is the nth root of the discretely compounded return.
$$R_t(n)_{avg} = \Pi(1 + R_{t-n})^\frac{1}{n} - 1$$

Further, market convention is to quote return on an annual basis so that investments can be compared. Because we likely have more or less than exactly one year of returns, we first need to find the average return and then compound that back out $p$ times, which is the number of periods in one year. These operations can be combined as:

$$R_t(n)_{ann} = \Pi(1 + R_{t-n})^\frac{p}{n} - 1$$

### Aggregating returns cross-sectionally
Given $i$ stocks held in a portfolio with weights $w_i$ equivalent to the dollar amount of stock held, the cross-sectional or portfolio return of the portfolio $R_p$ is the weighted average of the individual stock returns.

$$R_p = \Sigma{w_iR_i}$$

```python
# To illustrate, consider a 120k portfolio holding 20k of AAPL and 100k of IBM, starting in 2004
AAPL_pos = 20000
IBM_pos = 100000

# The cross-sectional portfolio total return is the weighted average of either the simple or total returns
portfolio_rtn = (AAPL_pos*returns['AAPL Total Rtn'] + IBM_pos*returns['IBM Total Rtn'])/(AAPL_pos+IBM_pos)
#portfolio_rtn_simple = (AAPL_pos*(returns['AAPL Total Rtn']-1) + IBM_pos*(returns['IBM Total Rtn']-1))/(AAPL_pos+IBM_pos)

# The (temporal) cumulative product of the total returns acts like an index for the portfolio value
((AAPL_pos+IBM_pos)*portfolio_rtn.cumprod()).plot(figsize=(14,5))
plt.title('Value of a 120k portfolio over time')
```

### The Lognormal Distribution

We have satisfied ourselves that stock returns meet our criteria for a market invariant. We have also seen that returns aggregate through time via multiplication. We need to tie these together with the random walk model, which recall is: 

$$(1+R_t)= \mu + \sigma\epsilon_t,\ \epsilon_t \sim IID\ N(0,1)$$

However, we encounter issues:

- If the variance term $\sigma$ is large enough, there is nothing stopping us from simulating a series of returns that aggregates to less than -100$\%$. This would imply a stock price of less than zero, which is inconsistent with actual market prices.
- The probability distribution of the product of $n$ normal random variables is not itself normal, making it difficult to model the conditional distribution of the market invariant at some future horizon $h$ (which is the distribution of the price forecast at $h$).

We'll instead model the total returns as a lognormally distributed random variable. The lognormal distribution is one where taking its natural log we get a normal distribution, hence the lognormal itself is exponential in nature.

$$(1+R_t) = \exp(\mu + \sigma\epsilon_t)$$

The lognormal helps us with both of the above issues: As it's exponential in nature, it cannot take negative values. The lognormal also naturally describes the product of multiple IID random variables, and the distribution of that product is itself lognormally distributed, making it tractable to model the future distribution of the price.  

Investigating the properties of the lognormal distribution, we have the PDF in the Python library SciPy. 

A lognormally distributed variable (i.e. one where log(X) has mean $\mu$ and variance $\sigma^2$) has parameters:
- 'shape' is $\sigma$, e.g. the std dev of log(X)
- 'loc' here is an offset that gets subtracted from the data, set this to 0.
- 'scale' is exp($\mu$)

```python
# Comparing the normal and lognormal distribution for our market invariant
logn_sigma, logn_loc, logn_exp_mu = lognorm.fit(returns['AAPL Total Rtn'], floc=0)
print logn_exp_mu, logn_sigma
norm_mu, norm_sigma = norm.fit(returns['AAPL Total Rtn'])
print norm_mu, norm_sigma
```

```python
plt.figure(figsize=(8,6))
x = np.linspace(.8, 1.2, num=150)
#x = np.linspace(.90, .95, num=150)  # left tail zoom - lognormal is smaller
#x = np.linspace(1.05, 1.10, num=150) # right tail zoom - lognormal is larger
plt.hist(returns['AAPL Total Rtn'], bins=100, alpha=.4, normed=True)
plt.plot(x, lognorm.pdf(x, logn_sigma, loc=0, scale=logn_exp_mu), color='r', label='lognormal model')
plt.plot(x, norm.pdf(x, loc=norm_mu, scale=norm_sigma), color='b', label='normal model')
plt.legend()
```

Fit to the daily returns, the lognormal looks similar to the normal. 

However, fit to the cumulative return over the entire time span being analyzed, the difference between the two distributions becomes apparent.

```python
# The price is the cumulative of the total returns, unscaled by the initial price P0
cumu_tot_rtns = px_adj_close['AAPL']/px_adj_close['AAPL'][0]
print np.allclose(returns['AAPL Total Rtn'].cumprod(), cumu_tot_rtns[1:]) #allclose checks equality

# Fit a lognormal and a normal to the total returns
logn_px_sigma, logn_px_loc, logn_px_exp_mu = lognorm.fit(returns['AAPL Total Rtn'].cumprod(), floc=0)
print logn_px_exp_mu, logn_px_sigma
norm_px_mu, norm_px_sigma = norm.fit(returns['AAPL Total Rtn'].cumprod())
print norm_px_mu, norm_px_sigma
```

```python
# The distribution of stock prices
fig = plt.figure(figsize=(10,6))

x = np.linspace(.05, 100, num=1000)
plt.hist(returns['AAPL Total Rtn'].cumprod(), bins=100, alpha=.4, normed=True)
plt.plot(x, lognorm.pdf(x, logn_px_sigma, loc=0, scale=logn_px_exp_mu), color='r', lw=2, label='lognormal model')
plt.plot(x, norm.pdf(x, loc=norm_px_mu, scale=norm_px_sigma), color='b', lw=2, label='normal model')
plt.legend()
```

The lognormal distribution captures the compound growth resulting from the multiplication of the individual total returns. The above shows the compound total return, which multiplied through by the initial price $P_0$ is the stock price $P_t$. 

#### Continuously Compounded or Log Returns

The lognormal distribution can be more difficult to parameterize and work with than the normal distribution. Fortunately, we don't need to work with it directly because we can simply take the natural log of our total returns, which lets us to work with a normal distribution (recall the log of the lognormal is a normal). 

Our market invariant is then the continuously compounded or log return, which we denote with a lower case $r$. Which gets us back to the original additive form of the random walk model.

$$r_t = ln(1 + R_t) = ln(\frac{P_t}{P_{t-1}}) = ln(P_t) - ln(P_{t-1}) = \mu + \sigma\epsilon_t,\ \epsilon_t \sim N(0,1)$$

#### Working with log returns

We know that discretely compounded returns aggregate temporally via multiplication and this results in exponential compound growth. Log returns represents a continuous rather than a discrete type of compound growth. For short time periods at low variance, the log return approximates the discretely compounded return and we can use them interchangeably for the purpose of risk calculations. However for longer time periods (which typically also have higher variance) the two values diverge. 

Log returns arise as the solution to the continuous time version of the random walk (Brownian motion) and they are the benchmark assumption when working in continuous time finance. 

Comparing the annual AAPL and IBM log vs simple returns:

```python
# Comparing discrete vs continuously compounded return

# the log returns
returns['AAPL Log Rtn'] = np.log(returns['AAPL Total Rtn'])
returns['IBM Log Rtn'] = np.log(returns['IBM Total Rtn'])

# The Pandas groupby chain first groups the cumulative return for each year, and then takes the last value from each group
ann_disc_rtns = returns[['AAPL Total Rtn','IBM Total Rtn']].groupby(lambda x: x.year) \
                 .cumprod().groupby(lambda x: x.year).last()-1  # -1 to transform simple rtns so comparable to log rtns
ann_disc_rtns.columns = [u'AAPL Simple Rtn', u'IBM Simple Rtn']

ann_cont_rtns = returns[['AAPL Log Rtn','IBM Log Rtn']].groupby(lambda x: x.year) \
                 .cumsum().groupby(lambda x: x.year).last()

ann_rtns = ann_disc_rtns.join(ann_cont_rtns)

# plot to compare
ann_rtns.plot(kind='bar', sort_columns=True, figsize=(12,6))

# format the yticks as percentages
new_labels = ['{:.0%}'.format(x) for x in np.arange(-1,2.6,.5)]
x_ = plt.gca().set_yticklabels(new_labels)
```

This example illustrates the difference in log vs simple returns for a long time period (annual) and a volatile stock (AAPL).
We note that the log return underestimates the true discrete return for years with positive total return, and overstates the losses in years with a negative total return.

If our goal is to compare the AAPL and IBM annual returns, we would use discretely compounded returns. However, given our goal is to find the best market invariant for a stock, we'll take advantage of the benefits of log returns and keep in mind the approximation they make.  

Though note that we can always transform the log return back to the simple return, even after taking the cumulative or the average. 

```python
# Transforms amongst log/simple/total returns
tspan = 252 # 1 year trailing window

print 'Cumulative Simple Return '
simple_cumu = (returns['AAPL Total Rtn'][-tspan+1:].cumprod()[-1]-1)
simple_single_period = (px_adj_close['AAPL'][-tspan:][-1]/px_adj_close['AAPL'][-tspan:][0]-1)
print '{:.2%}'.format(simple_cumu)
print '{:.2%}'.format(simple_single_period)

print '\nCumulative Log Return'
log_cumu = np.log(returns['AAPL Total Rtn'][-tspan+1:]).cumsum()[-1]
log_single_period = (np.log(px_adj_close['AAPL'][-tspan:][-1])-np.log(px_adj_close['AAPL'][-tspan:][0]))
print '{:.2%}'.format(log_cumu)
print '{:.2%}'.format(log_single_period)

print '\nTransform Cumulative Log Return to Simple Return'
print '{:.2%}'.format(np.exp(log_cumu)-1)

# what about the average
print '\nFind the Average Returns'
simple_avg = (1+simple_cumu)**(1./(tspan-1))-1 # geom avg for the simple
log_avg = log_cumu/(tspan-1) 
print 'Avg Simple Rtn {:.5%}'.format(simple_avg)
print 'Avg Log    Rtn {:.5%}'.format(log_avg)

print '\nTransform Cumulative Log Average to Simple Average'
print 'Avg Simple Rtn {:.5%}'.format(np.exp(log_avg)-1)
```

### Summary and Next Steps

In summary:

- Stock prices are not the IID market invariant we're trying to identify for our model. The price has a lag 1 dependence where the price at $t$ is heavily dependent on that at $t-1$. 
- The random walk models the evolution of the price as the cumulative of IID stock price changes which are samples from a standard normal distribution. 
- The random walk with drift adds in a constant per unit increase $\mu$ to the stock price process. 
- Stock price differences are independent, but are not identically distributed because they are sensitive to the absolute price level of the stock at time $t$.
- Percentage returns are independent and more identical than price differences.
- Non-constant variance $\sigma_t$ is a known feature of financial returns, which are thus not strictly IID. 
- Returns aggregate across time via multiplication and across portfolios via averaging.
- We observe that stock prices are best represented by the lognormal distribution, which prevents them from becoming negative and also models their compound growth. 
- Rather than work with the lognormal distribution and discrete returns, we prefer to work with the normal distribution and log returns $r_t = ln(1 + R_t)$
- The log return approximates the discrete return for short time periods and low variance.
